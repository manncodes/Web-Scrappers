```

for i in range(97,123):
  stime = time.time()
  print("doing->",str(chr(i)))
  URL ="https://www.webmd.com/a-to-z-guides/health-topics?pg="+str(chr(i))
  r=requests.get(URL)
  soup = BeautifulSoup(r.content, 'html5lib')
  #print(soup.prettify())
  ul = soup.find("ul",{"class":"az-index-results-group-list"})
  a_link_list=[]
  for li in ul.findAll("li"):
    a_link_list.append(li.a['href'])
  #print(a_link_list)
  for a_link in a_link_list:
    start_time=time.time()
    URL2 =a_link+"#2"
    try:
      r2=requests.get(URL2)
    except requests.RequestException:
      continue
    soup2 = BeautifulSoup(r2.content, 'html5lib')
    #print(soup2.prettify())
    sec = soup2.findAll("section")
    #2-symptom mostly, 3-cause
    try:
      list = sec[2].findAll("ul")
    except IndexError:
      list = []
    for ul in list:
      symp=[]
      for li in ul.findAll("li"):
        symp.append(li.text.strip())
    try:
      name = soup2.article.header.text.strip().replace("What Is ","")
    except AttributeError:
      name=""
    datacell={
        "name":name,
        "Symptom":symp
    }
    output = output.append(datacell, ignore_index=True)
    end_time=time.time()
    print(name," ",end_time-start_time)
  etime = time.time()
  print("Alphabet time:",etime - stime)


 
```

```
output.to_csv("disease2symptoms.csv")
```
